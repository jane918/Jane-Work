This is the analysis for the Churn data set, and the overall goal is to build a model to predict which customers leave and which customers stay for a fictional telecommunications firm. This note will start with EDA first, and build three binary classification models and evaluate their performance. 

```python
# Import the module needed 

# Data frame
import pandas as pd
import numpy as np

# sci-kit learn
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import (train_test_split, cross_val_score)
from sklearn.feature_selection import RFE
from sklearn.ensemble import (RandomForestClassifier,GradientBoostingClassifier)
from sklearn import preprocessing
from sklearn.metrics import (confusion_matrix, classification_report, roc_auc_score, 
                            roc_curve, auc)

# Data visualization
import matplotlib.pyplot as plt
import seaborn as sns  
```

```python
# Read the data
churn = pd.read_csv("churn.csv")

# explore the data 
print(churn.head(10))
# number of observations
print(churn.shape[0])
# number of variablesz
print(churn.shape[1])

churn.info()
```

We can see that there are twelve variables and 20,000 observations, and we will briefly summarize the type of variables as below. There is no missing data according to the information of dataframe, so we don't need to worry about the potential problems caused by missingness.

### Predictors
- Catergorical(Binary):College; 
- Quantitative: Income, Overage, Leftover, House, Handset Price, Over 15mins Calls Per Month, Average Call Duration.
- Catergorical(Multi levels): Reported Satisfication, Reported Usage Level, Considering Change of Plan. 

### Response Variable
- Leave - Catergorical(Binary): Whether the customer leaves or stays or not?

### Correlated Numerical Variables
Next, we will focus on the correlation between the numerical variables, from the correlation matrix, we know that there are three highly correlated pairs of variables, which is 'handset price' and 'income', 'over 15mins calls per month' and 'overage', and 'average call duration' and 'leftover'. All the three relationships make sense: if customer has higher income, he or she can afford higher handset price; if there is more overage, there will be more over 15mins call per month; if there is more leftover, there will be less average call duration. But the absolute value of pearson correlation coefficient is about 0.7(not very high), we will still keep those variables in our future models.  

```python
# Exploratory data analysis begins here

# Correlation Plot
data = churn.corr()
plt.figure(figsize=(12,8))
plt.title('Pearson Correlation of  Churn Features',y=1.05,size=15)
sns.heatmap(data, linewidths=0.05,vmax=1.0,
            square=True,linecolor='white',annot=True)  
plt.show()
```

### Univariate Data Visualization
Next, we want to examine how does each variable influence the response variable - LEAVE (with two levels 'LEAVE' or 'STAY'). For the categorical variable, we are going to draw bar plot against the response variable; for the quantitative variable, we are going to draw histograms. In total we have 11 predictors, and the univariate graphical EDA is as below. 

From these plots, we can see that for different levels of 'LEAVE', the distributions of 'income', 'overage', 'house', 'Over 15mins Calls Per Month', and 'Reported Satisfication' are significantly different. For example, the customers who leave have higher income on average compared with the customers who stay, maybe people with higher income can afford the additional fee if they change a new fictional telecommunications firm. And constumers with higher reported satisfication level are more likely to stay, which corresponds with intution. 

```python
# The univariate EDA (regarding 'LEAVE')

# split the data set by 'LEAVE' or 'STAY'
churn_leave = churn[churn['LEAVE']=="LEAVE"]
churn_stay = churn[churn['LEAVE']=="STAY"]

cat_variable =["COLLEGE", "REPORTED_SATISFACTION", "REPORTED_USAGE_LEVEL", 
               "CONSIDERING_CHANGE_OF_PLAN"]
num_variable = ["INCOME", "OVERAGE", "LEFTOVER", "HOUSE", "HANDSET_PRICE", "OVER_15MINS_CALLS_PER_MONTH",
                "AVERAGE_CALL_DURATION"]

# Numerical Variable - Density
plt.figure(figsize=(14,10))
plt.suptitle("Quantitative Variables vs LEAVE/STAY",fontsize=16,x=0.53,y=0.95)
for i in range(len(num_variable)):
    plt.subplot(int('33'+str(i+1)))
    sns.distplot(churn_leave[num_variable[i]] , color="maroon", label="Leave", hist=False)
    sns.distplot(churn_stay[num_variable[i]] , color="indianred", label="Stay", hist=False)
    
# Categorical Variable - Barplot
plt.figure(figsize=(12,8))
plt.suptitle("Categorical Variables vs LEAVE/STAY",fontsize=16,x=0.53,y=0.95)
for i in range(len(cat_variable)):
    plt.subplot(int('22'+str(i+1)))
    sns.countplot(x=cat_variable[i], hue ="LEAVE", data=churn, palette="rocket")
```


### Split Training Set and Test Set 

Next, we will build several classification models to predict "LEAVE". Since the scikit-learn package cannnot take in string, we will need the 'LabelEncoder' to transform the categorical variables into integers. And to prevent over-fitting, we will also split the data set into training set(80%) and test set(20%). 

```python
# change the categorical variables
le = preprocessing.LabelEncoder()
labelchurn = churn.apply(le.fit_transform)

# Divide the original data set into predictors and response variable
churn_X = labelchurn.drop(columns=['LEAVE'])
churn_y = labelchurn["LEAVE"]

# Generate training set and test set
X_train, X_test, y_train, y_test = train_test_split(churn_X, churn_y, test_size=0.2)
```

### Logistic Model

For the first model, the training set will be fixed using logistic regression, which is a commonly used binary classification model. Overall the accuracy is 0.63225, and the f1 score for label 'LEAVE' is 0.63, and for label 'STAY' is 0.63. This is not very high, so we will look into other advanced models. 

We also draw the predicted probability plot about whether the customer will stay. And we can see that the proportion of predicted probability which is bigger than 0.5 is higher than the proportion of probability less than 0.5, so if we set the threshold as 0.5, more customers will be predicted as 'STAY'. Meanwhile, lots of predicted probabilities fall in the range of 0.45 to 0.7. If we change the the threshold in this range, the predicted label of 'LEAVE' for many customers will change consequently.

```python
# Build the first binary model

# Logistic regression
lgmodel = LogisticRegression(penalty='l2', solver='liblinear').fit(X_train, y_train)
# Returns the mean accuracy on the given test data and labels.
score = lgmodel.score(X_test, y_test)
print(score)

# Make prediction on the test set
lgpredict = lgmodel.predict(X_test)
lg_confusion_matrix = confusion_matrix(lgpredict, y_test)
print(lg_confusion_matrix)
print(classification_report(y_test, lgpredict))

# store the predicted probabilities for class 1 (stay)
y_pred_prob = lgmodel.predict_proba(X_test)[:, 1]

plt.figure(figsize=(10,6))
sns.distplot(y_pred_prob, color="indianred", hist=False)
plt.xlim(0,1)
plt.title('Histogram of logistic predicted probabilities for "STAY" ')
plt.xlabel('Predicted probability of STAY')
plt.ylabel('Frequency')
```

### Classification Tree and Random Forest
    
    
Then we will consider building a random forest to make the prediction, which is a freqently used bagging method. Before building the random forest, we will use cross validation to tune the parameters of 'number of trees in the forest' and 'max depth of a single tree', and eight combinations will be tried. After that, we find the best combination is 200 and 4. 

From the confusion matrix and statistical report, we can see that the accuracy is about 0.69, and the f1 score for label 'LEAVE' is 0.67, and for label 'STAY' is 0.71, which means that the performance of random forest is better than the logistic model, and overall the model's performance of predicting 'STAY' is greater than predicting 'LEAVE'. But the precision of predicting 'LEAVE' is higher than predicting 'STAY', so it depends on what metric we want to use to evaluate the model.

```python
# Decision tree and random forest

# Use cross-validation to tune parameters 
# There will be 8 combinations 
# We will choose the combination with the smallest error
n_estimators = [50, 100, 150, 200]
max_depth = [2, 3, 4]
max_score = 0
for i in range(len(n_estimators)):
    for j in range(len(max_depth)):
        rf = RandomForestClassifier(n_estimators=n_estimators[i], max_depth=max_depth[j],
                             random_state=0)
        scores = cross_val_score(rf, churn_X, churn_y, cv=5).mean()
        if scores > max_score:
            max_score = scores
            best_parameter = [n_estimators[i], max_depth[j]]

# This is the combination of 'number of estimators' and 'max_depth' for a random forest
print(best_parameter)

# fit the tree
rf = RandomForestClassifier(n_estimators=best_parameter[0], max_depth=best_parameter[1],
                             random_state=0).fit(X_train, y_train)
print(max_score)

print(rf.score(X_test, y_test))
rfpredict = rf.predict(X_test)
rf_confusion_matrix = confusion_matrix(rfpredict, y_test)
print(rf_confusion_matrix)
print(classification_report(y_test, rfpredict))
```

### Feature Importance by Random Forest

In the random forest, the features that are frequently choosen as the split points will be more important predictors than the others. From the feature ranking, we can see the importance for each feature, and we can also see the rank more clearly from the feature importances plot.

We can find that feature 4-HOUSE and 2-OVERAGE are very important, and feature 10-CONSIDERING_CHANGE_OF_PLAN, 8-REPORTED_SATISFACTION, 9-REPORTED_USAGE_LEVEL and 0-COLLEGE are not important, which is consistent with our EDA. 

It is a little suprising that 'REPORTED_SATISFACTION' is not important, but it is reasonable because we can see from the barplot that the difference of 'LEAVE' or 'STAY' for each level of reported satisfaction is not very large. 

```python
importances = rf.feature_importances_

std = np.std([tree.feature_importances_ for tree in rf.estimators_],
             axis=0)
indices = np.argsort(importances)[::-1]

# Print the feature ranking
print("Feature ranking:")

for f in range(X_train.shape[1]):
    print("%d. feature %d (%f)" % (f + 1, indices[f], importances[indices[f]]))

# Plot the feature importances of the forest
plt.figure(figsize=(12,8))
plt.title("Feature importances by Random Forest")
plt.bar(range(X_train.shape[1]), importances[indices],
       color="indianred", yerr=std[indices], align="center")
plt.xticks(range(X_train.shape[1]), indices)
plt.xlim([-1, X_train.shape[1]])
plt.show()
```

### Gradient Boosting Tree

Next, we will use gradient boosting tree to compare the perfomance of random forest, which is another kind of ensemble learning algorithm by building a series of boosting trees. Similarly, we will use cross validation to tune the parameters of 'number of trees' and 'max depth of a single tree', and eight combinations will be tried. After that, we find the best combination is 50 and 2. 


From the confusion matrix and statistical report, we can see that the accuracy is about 0.69, and the f1 score for label 'LEAVE' is 0.70, and for label 'STAY' is 0.68. The difference of model's performance predicting 'STAY' or 'LEAVE' is smaller compared with the random forest. Overall I think the gradient boosting tree has very similar performance with the random forest, but maybe it is slightly better because the F1-scores for two labels are closer to each other. 

```python
# Gradient Boosting Tree

# Use cross-validation to tune parameters 
# There will be 8 combinations 
# We will choose the combination with the smallest error
n_estimators = [50, 100, 150, 200]
max_depth = [2, 3, 4]
max_score = 0
for i in range(len(n_estimators)):
    for j in range(len(max_depth)):
        gbt = GradientBoostingClassifier(n_estimators=n_estimators[i], max_depth=max_depth[j],
                             random_state=0, learning_rate=1.0)
        scores = cross_val_score(gbt, churn_X, churn_y, cv=5).mean()
        if scores > max_score:
            max_score = scores
            best_parameter = [n_estimators[i], max_depth[j]]

# This is the combination of 'number of estimators' and 'max_depth' for the gradient boosting trees
print(best_parameter)


gbt = GradientBoostingClassifier(n_estimators=best_parameter[0], max_depth=best_parameter[1],
                             random_state=0, learning_rate=1.0).fit(X_train, y_train)
print(max_score)


print(gbt.score(X_test, y_test))
gbtpredict = gbt.predict(X_test)
gbt_confusion_matrix = confusion_matrix(gbtpredict, y_test)
print(gbt_confusion_matrix)
print(classification_report(y_test, gbtpredict))
```

### Feature Importance by Gradient Boosting Tree

For the gradient boosting tree, we can see that only feature 2, 4, 1, 3 are important, which are INCOME, OVERAGE, LEFTOVER and HOUSE.

```python
importances = gbt.feature_importances_
print("Feature ranking:")
indices = np.argsort(importances)[::-1]
for f in range(X_train.shape[1]):
    print("%d. feature %d (%f)" % (f + 1, indices[f], importances[indices[f]]))
```

### Model Comparision

```python
# access and compare model's performance

# the base rate (for naive classifier)
print("Naive Rate:", sum(churn["LEAVE"] == "STAY")/20000)

# ROC curve
# Logistic
lgfpr,lgtpr,lgthreshold = roc_curve(y_test, lgpredict) 
lgroc_auc = auc(lgfpr,lgtpr)

# Random Forest
rffpr, rftpr, rfthreshold = roc_curve(y_test, rfpredict) 
rfroc_auc = auc(rffpr,rftpr)

# Gradient Boosting Tree
gbtfpr, gbttpr, gbtthreshold = roc_curve(y_test, gbtpredict) 
gbtroc_auc = auc(gbtfpr,gbttpr)

plt.figure()
lw = 2
plt.figure(figsize=(10,10))
plt.plot(lgfpr, lgtpr, color='maroon',
         lw=lw, label='ROC curve of logistic regression model (area = %0.2f)' % lgroc_auc)
plt.plot(rffpr, rftpr, color='indianred',
         lw=lw, label='ROC curve of random forest (area = %0.2f)' % rfroc_auc)
plt.plot(gbtfpr, gbttpr, color='pink',
         lw=lw, label='ROC curve of gradient boosting tree (area = %0.2f)' % gbtroc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC for several classifiers')
plt.legend(loc="lower right")
plt.show()
```

All the models have better performance compared with the naive classifiers. From the ROC curve we think random forest and gradient boosting tree have the similiar performance because they have the same AUC. And from the inflection point, we should try to set the threshold(p) to make TP rate be about 0.75 for the random forest, instead of just choosing p=0.5. 

### Summary and Next Step
